```python
import numpy as np
import matplotlib.pyplot as plt
import pyvista as pv
```

```python
pts = np.loadtxt("psinew.dat", unpack = True)
```

```python
N, r_avg, r_err = np.loadtxt("r_avg.dat", unpack = True)
N, r_avg_1s, r_err_1s = np.loadtxt("r_avg_1s.dat", unpack = True)
```

```python
plt.figure(figsize = (10,5))
plt.axhline(5, color = "red", label = r"$\langle r\rangle_{analytic}$ ")
plt.errorbar(N, r_avg, r_err, color = "blue", linewidth =1, elinewidth = 0.5,ecolor = "blue", label = r"$\langle r\rangle_{M(RT)^2}$ ")
plt.legend()
plt.title(r"Estimation of $\langle r \rangle$ for the $\Psi_{2,1,0}$ orbital of the Hydrogen atom")
plt.xticks(np.arange(0,110, 10))
plt.grid()
plt.xlabel("Block number")
plt.ylabel(r"r [$a_0$]")
plt.ylim((4.95,5.07))
plt.xlim((0, 100))
```

```python
plt.figure(figsize = (10,5))
plt.axhline(1.5, color = "red", label = r"$\langle r\rangle_{analytic}$ ")
plt.errorbar(N, r_avg_1s, r_err_1s, color = "blue", linewidth =1, elinewidth = 0.5,ecolor = "blue", label = r"$\langle r\rangle_{M(RT)^2}$ ")
plt.legend()
plt.title(r"Estimation of $\langle r \rangle$ for the $\Psi_{1,0,0}$ orbital of the Hydrogen atom")
plt.xticks(np.arange(0,110, 10))
plt.grid()
plt.xlabel("Block number")
plt.ylabel(r"r [$a_0$]")
plt.xlim((0, 100))
```
