```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1s Orbital

For the 1s orbital it was found that a maximum step of 1.23 gives the desired 50% acceptance rate. Given the spherical symmetry of the 1s orbital the starting point was set at 
$$ \begin{align}
    \vec{r} &= \begin{bmatrix}
           1 \\
           1 \\
           1
         \end{bmatrix}
  \end{align}a_0
$$
this gives an average radius of  $ \langle r \rangle = 1.50118 a_0 \pm 0.0053746 a_0$

Below the results from the simulation are reported as a plot comparing the expecation value of the radius obtained analytically 
$$
    \langle r \rangle = \frac{3}{2} a_0
$$


```python
N, r_avg, r_err = np.loadtxt("../../src/lab5/rAvg_1s.dat", unpack = True)
```

```python
plt.figure(figsize = (8,3))
plt.plot(N, r_avg, c = "b", label = r"$\langle r\rangle_{MC}$ ")
plt.axhline(1.5, color = "red", linestyle = "--",label = r"$\langle r\rangle_{analytic}$ ")

plt.fill_between(N, r_avg-r_err, r_avg+r_err, alpha = 0.4, edgecolor ="b", facecolor = "b")
plt.legend()
plt.title(r"Estimation of $\langle r \rangle$ for the $\Psi_{1,0,0}$ orbital of the Hydrogen atom")
plt.xticks(np.arange(0,110, 10))
plt.grid()
plt.xlabel("Block number")
plt.ylabel(r"r [$a_0$]")
plt.ylim((1.4, 1.55))
plt.xlim((0, 100))
```

## 2p Orbital

For the 1s orbital it was found that a maximum step of $2.99 a_0$ gives the desired 50% acceptance rate. Given the elongated along the z axis shape of the 2p orbital the starting point was set at

## Orbital density plot

To have a feel for the validity of the results the output from the position sampling is plotted as a 3d scatter plot which gives a density plot of the electornic cloud, which, for the results to be valid, should resemble the shape of the orbital

### 1s Orbital

Below the density plot for the 1s orbital shows its spherical symmetry as expected

```python
x_1, y_1, z_1 = np.loadtxt("../../src/lab5/psi2_1s.dat", unpack = True)
r_1 = []
for i in range (len(x_1)):
    r = x_1[i] * x_1[i] + y_1[i] * y_1[i] + z_1[i] * z_1[i] 
    r = np.sqrt(r)
    r_1.append(r)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x_1, y_1, z_1, c=r_1, marker='.', s = 2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)

plt.show()
```
