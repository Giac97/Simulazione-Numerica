## Metropolis

Then the values obtained for the measured quantities (u, c, m and $\chi$) are loaded from the files generated using the code

### Internal Energy

We first load the data from the simulation with the results for the internal energy

```python
T, block, est_u, u_av, u_err = np.loadtxt("energy.dat", unpack = True)
```

The file contains in the first column the temperature, the second column is the number of the final block, the following columns contain the estimated value, the block averaged value and the uncertainty

In order to check the validity of the result we plot it together with the exact value obtained from the analytical formula:

To do so we first define an array for a temperature ranging from the minumum to the maximum value of the temperature used in the simulation. As the temperature range is the same for all simulations we can do this only once, we also evaluate the inverse of the temperature (the $\beta$ parameter), the interaction constant and the number of spins. This values will be used to evaluate the theoretical values to be compared with the results from the simulation.

```python
T_th = np.linspace(min(T), max(T), 100)
beta = 1. / T_th
J = 1.0
Ns = 50
```

```python
th = np.tanh(J/T_th)
thN= th**Ns

ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
```

```python
plt.errorbar(T, u_av, u_err,  ls ='', elinewidth = 0.5, marker = "*", markersize = 0.7,label = r"$U_{Monte Carlo}$")
plt.plot(T_th, e, linewidth = 0.5, color = "red", label = r"$U_{Theory}$")
plt.xlabel("T")
plt.ylabel("U")
plt.title(r"U Vs. T theoretical and MC (Metropolis) value", fontsize = 15)
plt.legend()
```

### Specific Heat

```python
T, block, est_c, c_av, c_err = np.loadtxt("heat.dat", unpack = True)
```

As with the internal energy we compare the results from the simulation with the results from the analytical solution of the 1d ising model

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.figure(figsize = (6, 4))
plt.errorbar(T, c_av, c_err, ls ='', elinewidth = 0.5, marker = "*", markersize = 0.7,label = r"$C_{Monte Carlo}$")
plt.plot(T_th, heat, linewidth = 0.5, color = "red", label = r"$C_{Theory}$")
plt.xlabel("T")
plt.ylabel("C")
plt.title(r"C Vs. T theoretical and MC (Metropolis) value", fontsize = 15)
plt.xticks(np.arange(0.4, 2.1, 0.2))
plt.legend()
```

### Magnetization

We then load the data for the magnetization, note that this data was obtained from a separate simulation with a positive value for the external field of $h = 0.02$

```python
T, block, est_m, m_av, m_err = np.loadtxt("magnet.dat", unpack = True)
```

As before the data from the simulation are compared with the analytical values

```python
h=0.02
l1 = np.exp(beta*J)*np.cosh(beta*h)+np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))
l2 = np.exp(beta*J)*np.cosh(beta*h)-np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))
Z = l1**Ns + l2**Ns
M = (np.exp(beta*J)*np.sinh(beta*h)*((l1**(Ns-1))*(1+np.exp(beta*J)*np.cosh(beta*h)/np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))) 
        + (l2**(Ns-1))*(1-np.exp(beta*J)*np.cosh(beta*h)/np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J)))))/(Z)
```

```python
plt.figure(figsize = (6, 4))
plt.errorbar(T, m_av, m_err, ls ='', elinewidth = 0.5, marker = "*", markersize = 0.7,label = r"$m_{Monte Carlo}$" )
plt.plot(T_th, M, linewidth = 0.5, color = "red", label = r"$m_{Theory}$")
plt.title(r"M Vs. T theoretical and MC (Metropolis) value", fontsize = 15)
plt.xlabel("T")
plt.ylabel("m")
plt.legend()
```

For this simulation we will also have a look at the final configurations reached for every simulation, we first generate an array filled with zeros with size $N_{spins} \times N_{temp}$, where $ N_{temp}$ is the number of temperature sweeps

```python
config = np.zeros((50,len(T)))
```

then for each temperature we read the configuration from the output configuration of the simulation:

```python
for i in range(len(T)):
    fname = "./configs/config.final."
    tstring = str(T[i])
    fname+=tstring
    config[:,i] = np.loadtxt(fname, unpack = True)
```

The result is then plotted with imshow, each column is a configuration at a given temperature, going from $T_{min}$ on the left to $T_{max}$ on the right, here with represent a spin up, black a spin down:

```python
plt.imshow(config, cmap="gray", interpolation='none', extent=[min(T),max(T),0,5], aspect=0.5)
plt.xlabel("T")
plt.xticks([0.4, 1.0, 1.5, 2.0])
```

```python
T, block, est_x, x_av, x_err = np.loadtxt("sus.dat", unpack = True)
```

```python
X_th = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
```
