### Solid
In order to obtain a constant temperature of $T^* = 0.8$ a starting temperature of $T^* = 1.6$ was found to be optimal. The simulation was run for a total of 50,000 steps divided in 50 blocks of 1,000 steps each.

We first load the output data from the simulation by reading the content of the solid folder

```python
solid_path = "./solid/"
solid_outputs = os.listdir(solid_path)
```

```python
print(solid_path+solid_outputs[0])
```

then the data is assigned to repsective arrays

```python
blck, ekin, ekinblck, ekinerr = np.loadtxt(solid_path+solid_outputs[0], unpack = True)
blck, epot, epotblck, epoterr = np.loadtxt(solid_path+solid_outputs[1], unpack = True)
blck, etot, etotblck, etoterr = np.loadtxt(solid_path+solid_outputs[2], unpack = True)
blck, press, pressblck, presserr = np.loadtxt(solid_path+solid_outputs[3], unpack = True)
blck, temp, tempblck, temperr = np.loadtxt(solid_path+solid_outputs[4], unpack = True)
```

First a plot of the energies as function of the number of blocks is shown

```python
fig, axs = plt.subplots(3, sharex = True, figsize = (8, 5))
axs[0].plot(blck, ekinblck, c = "r")
axs[0].fill_between(blck, ekinblck-ekinerr, ekinblck+ekinerr, alpha = 0.4, edgecolor ="r", facecolor = "r")
axs[0].set_xlim((0,50))
axs[0].set_ylabel(r"$E_{kin}*$", fontsize = 14)

axs[1].plot(blck, epotblck, c = "b")
axs[1].fill_between(blck, epotblck-epoterr, epotblck+epoterr, alpha = 0.4, edgecolor ="b", facecolor = "b")
axs[1].set_xlim((0,50))
axs[1].set_ylabel(r"$E_{pot}*$", fontsize = 14)

axs[2].plot(blck, etotblck, c = "y")
axs[2].fill_between(blck, etotblck-etoterr, etotblck+etoterr, alpha = 0.4, edgecolor ="y", facecolor = "y")
axs[1].set_xlim((0,50))
axs[2].set_ylabel(r"$E_{tot}*$", fontsize = 14)
axs[2].set_xlabel("Block")
```

Then a plot for the temperature and pressure

```python
fig, axs = plt.subplots(2, sharex = True, figsize = (8, 5))
axs[0].plot(blck, pressblck, c = "b")
axs[0].fill_between(blck, pressblck-presserr, pressblck+presserr, alpha = 0.4, edgecolor ="b", facecolor = "b")
axs[0].set_xlim((0,50))
axs[0].set_ylabel(r"$P*$", fontsize = 14)

axs[1].plot(blck, tempblck, c = "r")
axs[1].fill_between(blck, tempblck-temperr, tempblck+temperr, alpha = 0.4, edgecolor ="r", facecolor = "r")
axs[1].set_xlim((0,50))
axs[1].set_ylabel(r"$T*$", fontsize = 14)
```
