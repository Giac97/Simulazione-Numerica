```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def psi_trial(x, sigma, mu):
    exp1 = -(x - mu) ** 2 / (2 * sigma ** 2)
    exp2 = -(x + mu) ** 2 / (2 * sigma ** 2)
    return np.exp(exp1) + np.exp(exp2)
```

### Sampling energy after Annealing
After completing the simulated annealing run, the system was sampled at constant temperature to find the estimate of $\sigma$ and $\mu$ more accurately and with their error estimate. Then the simAnneal flag was set to false in order to sample the wavefunction and the energy and its accuracy.

Below a plot of the energy $\langle H \rangle$, sampled after the annealing run, is shown with the error, compared with the solution obtained from matrix diagonalization. It can be seen that, while the energy does reach a convergence value of $\langle H \rangle = -0.4393 \pm 0.0029$, this value is off from the value found by diagonalization: $\langle H \rangle =-0.4605$. $\sigma = -0.6173 +/- 0.0002$

We now produce a plot of the sampled wavefunction and compare it with the analytical solution obtained from the eigenvalue equation

```python
position_anneal = np.loadtxt("./annealout/position.out")
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
```
