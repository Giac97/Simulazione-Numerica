```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercise 3.1
### Direct pricing

Here we do our Monte Carlo simulation evaluating directly the market price at the delivery time

```python
N, callPriceDir, callErrDir, putPriceDir, putErrDir = np.loadtxt("direct_pricing.dat", unpack = True) 
```

```python
fig, axs =  plt.subplots(1,2, figsize = (10, 4), dpi = 150)
gs = fig.add_gridspec(2,hspace=15)
axs[0].axhline(14.98, color = "red", linewidth = 0.7, label = "B-S Solution")
axs[0].errorbar(N, callPriceDir, callErrDir, ecolor = "black", elinewidth = 0.5, label = "MC Solution")
axs[0].legend()
axs[0].set_title("Call Option Price")
axs[0].set_xticks(np.arange(0, 110, 10))
axs[0].set_xlabel("N")
axs[0].set_ylabel("C[S(0), 0]")
axs[0].grid()

axs[1].axhline(5.4595325819072364, color = "red", linewidth = 0.7, label = "B-S Solution")
axs[1].errorbar(N, putPriceDir, putErrDir, ecolor = "black", elinewidth = 0.5, label = "MC Solution")
axs[1].legend()
axs[1].set_title("Put Option Price")
axs[1].set_xticks(np.arange(0, 110, 10))
axs[1].set_xlabel("N")
axs[1].set_ylabel("P[S(0), 0]")
axs[1].grid()
fig.savefig("direct_pricing.png")
```
